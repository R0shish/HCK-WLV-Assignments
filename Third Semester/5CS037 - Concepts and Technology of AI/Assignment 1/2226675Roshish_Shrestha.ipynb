##**1.2** Data Transformation:

Write code to transform variables according to the following instructions:

1. â€Schoolâ€, â€sexâ€, â€addressâ€, â€schoolsupâ€, â€famsupâ€, â€activitesâ€, â€nurseryâ€, â€internetâ€, and â€romanticâ€
into binary: 0 or 1 (create new columns without overwriting the existing ones).


```python
# A list of given columns
column_binary = ["school", "sex", "address", "schoolsup", "famsup", "activities", "nursery", "internet", "romantic"]

# Mapping through the given column and creating a new column with 
# "Yes to 1 and No to 0", "GP to 0 and MS to 1", "M to 0 and F to 1", "U to 0 and R to 1"
for column in column_binary:
    df[column+"_binary"] = df[column].map({"yes":1, "no":0, "GP":0, "MS":1, "M":0, "F":1, "U":0,"R":1})
df
```

2. â€Meduâ€, â€Feduâ€, â€reasonâ€, â€guardianâ€, â€studytimeâ€, â€freetimeâ€, and â€healthâ€ into ordinal numbers based
on the number cases in the data set (create news columns without overwriting the existing ones).

```python
# A list of given columns
columns_ordinal = ['Medu', 'Fedu', 'reason', 'guardian', 'studytime', 'freetime', 'health']

# Looping through given columns and encoding into ordinal numbers
for column in columns_ordinal:
    df[column + '_ordinal'] = pd.Categorical(df[column]).codes
df
```

3. Convert column â€ageâ€ to interval datatype. i.e. Create a new column name category age whose values
should be based on the frequency in the column â€ageâ€, You can create categorical data with following
interval.

  (a) interval1: [15-17]

  (b) interval2: [18-20]

  (c) interval3: [21-all]

```python
# Define the given intervals for the categories_age column
intervals = [15, 17, 20, np.inf]

# Create the age categories using pd.cut and include lowest to take lower bound into account
df['category_age'] = pd.cut(df['age'], intervals, labels=['[15-17]', '[17-20]', '[20-all]'], include_lowest=True)

df
```

4. Create a new column name passed (yes or no) whose values should be based on the values present in the
G3 column (>= 8â€“yes, < âˆ’no).

```python
# Using pandas.Series.apply() and passing an anonymous function that checks if G3 
# is greater than and equal to 8 or not and returns "yes/no" accordingly.
df['passed'] = df['G3'].apply(lambda x: "yes" if x >=8 else "no")
df
```

2. Write a code to show a bar graph of the total number of students who passed the final term grouped
according to the school that they belong to. Use proper labels in the graph and interpret the results.

```python
# By using the previously made "Passed" column
# count the number of students who passed and group them by their schools
school_counts = df.groupby(['school'])['passed'].apply(lambda x: x[x == 'yes'].count())

# Plot a bar graph with the count
school_counts.plot(kind='bar')

# Labeling the bar chart
plt.title('Number of Students Who Passed Final Term by School')
plt.xlabel('Schools')
plt.ylabel('Number of Students Who Passed Final Term')

# Show the plot
plt.show()
```

**Answer**:

The above graph shows the number of students who passed final term by school. Here, the x-axis represents the school and y-axis represents the number of students who passed final term. The graph depicts the frequency of students who passed final term from each school. We can clearly see that there are many graduates of school GP than MS.

3. Write a code to show a pie chart of analysis of studentsâ€™ grades as per their school. Use proper labels
in the graph and interpret the results.

```python
# Group the data by school
grouped_df = df.groupby('school')

# Create a pie chart for each school
for name, group in grouped_df:
    group['G3'].value_counts().plot.pie(title='Student Grades for ' + name, figsize=(11,11))
    plt.legend(title='Grades', loc='upper right', fontsize=8)
    plt.show()
```

**Answer**:
1. The pie-chart shows the distribution of final grades from 0 till 20 of school GP. We can clearly see that the amount of students who got 10 is maximum where as the grade 4 is minimum.

2. The pie-chart shows the distribution of final grades of school MS from 0 till 20. We can clearly see that the amount of students who got 10 is maximum where as the grade 6,14 and 19 is equal and is minimum. 

The marks are more evenly distributed in school MS than in GP.

4. Write a code to show a bar graph with the relation of the fatherâ€™s occupation with the grade of the
student. Use proper labels in the graph and interpret the results.

```python
# Group the data by father's occupation
# and Calculate the mean grade for each occupation
mean_grades = df.groupby("Fjob")["G3"].mean()

# Plotting the bar graph
mean_grades.plot.bar(title="Relation of Father's Occupation with Grade")

# Labelling the graph
plt.xlabel("Father's Occupation")
plt.ylabel("Mean Grade")

plt.show()
```

Above is a graph which depicts the relation of Father's Occupation with Grade. Here, x-axis represents the father's occupation like staying at home, health sector etc and the y-axis represents the mean grade of students. The meaning behind the graph plot is to show how the job of the father affects the grade of the students. The mean grade of students grade is taken and is compared with the occupation.
