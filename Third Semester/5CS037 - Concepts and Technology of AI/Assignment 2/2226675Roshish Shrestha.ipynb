```python
# Convert the image to grayscale and displaying it
img_grey = img.convert('L')
img_grey
```

##**5.2**  Standardize/Scale the data:

```python
# Converting the grayscale image to numpy array
img_arr = np.array(img_grey)

# Calculating the mean and standard deviation of the image array
mean = img_arr.mean()
standard_deviation = img_arr.std()

# Standardizing the image array
img_standardized = (img_arr - mean) / standard_deviation
```

## **5.3** Calculate Covariance Matrix:

Each element of the covariance matrix represents covariance between each ijth element. The covariance between
two elements is calculated and stored in the matrix as shown in picture below:
# 5.3.JPG

```python
# Calculating the covariance matrix of the standardized image array
cov_mat = np.cov(img_standardized.T)
print('Covariance matrix: \n%s' %cov_mat)
```

##**5.4** Eigen Decomposition:

Decompose a covariance matrix into eigen values and eigen vectors:
# 5.4.JPG

**Where:**
1. A(n*n): −Covariance Matrix.
2. P:- Eigen Vector of Covariance Matrix.
3. D:- Diagonal Matrix. Diagonal elements are Eigen Values.
4. P^I: −Inverse of Eigen Vector.

```python
# Decomposing the covariance matrix into eigenvalues and eigenvectors
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors: \n%s' %eig_vecs)
print('\nEigenvalues: \n%s' %eig_vals)
```

##**5.5** Identify Principal Components:

Determine the **explained variance** by each **principal components**, based on this we can determine how many principal components to pick from.
*   Experiment with Principal Components:<br>
– Pick **three different combination** of principal components with various **explained variance value** and **compare** the result.

```python
# Sorting the eigen values in decending order
eig_vals_sorted = np.sort(eig_vals)[::-1]
# Calculating the explained variance for the sorted eigenvalues
explained_variance = eig_vals_sorted / eig_vals_sorted.sum() * 100

# Choosing three principal components
principal_component_choices = [20, 50, 80]

# Printing the explained variance of the chosen three principal components
for principal_component in principal_component_choices:
  explained_variance_n = explained_variance[:principal_component]
  print("Number of components:", principal_component)
  print("Explained variance:", explained_variance_n.sum())
```

**Answer:** The first 20 principal components capture 87.46% of the variance in the data, while the first 50 principal components capture 95.52% of the variance in the data and the first 80 principal components capture 97.80% of the variance in the data. As a result, using more principal components leads to more variance in the data being explained.





##**5.6** Reconstruction of the image:

Reconstruct the image based on number of principal components you picked in step **5.5**. Reconstruct all three images and conclude the result based on your observation. Display the reconstructed image, experiment with various number of components.

```python
# Function to reconstruct the image with the given number of principal components
def reconstruct(n):
  print("When principal Component = " + str(n), end="\n\n")

  # Calculating the image data using the given principal components by taking the dot product of img_standardized and eigen_vector.
  eigen_vector = eig_vecs[:,:n]
  image_data = np.dot(img_standardized, eigen_vector)

  # Reconstructing the image from the image data by taking the dot product of image_data and the transpose of eigen_vector
  # Adding .real at the end to only take the real part of the resulting matrix
  image_reconstructed = np.dot(image_data, eigen_vector.T).real

  fig, axs = plt.subplots(1, 3, figsize=(25, 20))

  # Displaying the original image
  axs[0].imshow(img)
  axs[0].set_title('Original Image')

  # Displaying the grayscale image
  axs[1].imshow(img_grey, cmap='gray')
  axs[1].set_title('Grayscale Image')

  # Displaying the reconstructed image
  axs[2].imshow(image_reconstructed, cmap='gray')
  axs[2].set_title('Reconstructed Image')
```

```python
# Reconstructing the image with principal component = 20
n = principal_component_choices[0]
reconstruct(n)
```

```python
# Reconstructing the image with principal component = 50
n = principal_component_choices[1]
reconstruct(n)
```

```python
# Reconstructing the image with principal component = 80
n = principal_component_choices[2]
reconstruct(n)
```

**Observation**: With Principal component analysis, we can preserve the information of the image and also reduce the dimensionality of the data. The number of principal components we use determines how much of the variance in the data is preserved.

Taking the example of the above three choices. Choosing 20 principal components, we get quite a distorted image as it captures less than 88% of the variance in the data. Choosing 50 principal component gives us a clearer image while choosing 80 principle component captures 97% of the variance in the data.

Thus, the more principal components we use, the more variance in the data will be explained and as a result clearer image will be constructed. The number of principal components to use should be chosen based on the desired level of dimensionality reduction and the amount of information that should be retained.
